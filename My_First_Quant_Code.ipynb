```python
import yfinance as yf
import matplotlib.pyplot as plt

# 1. تحديد رمز الشركة والفترة الزمنية
ticker = "AAPL"
# قمنا بتعطيل الفهرس متعدد المستويات لتجنب الأخطاء
data = yf.download(ticker, start="2025-01-01", end="2026-06-15", multi_level_index=False)

# 2. حساب العوائد اليومية باستخدام عمود 'Close' (الذي أصبح معدلاً تلقائياً)
data['Daily_Return'] = data['Close'].pct_change()

# 3. طباعة أحدث البيانات للتأكد من عمل الكود
print("تم سحب البيانات بنجاح! أحدث الأسعار المستخرجة:")
print(data['Close'].tail())

# 4. رسم حركة سعر السهم وإظهار المخطط التفاعلي
plt.figure(figsize=(10, 5))
plt.plot(data['Close'], label='Apple Stock Price (Adjusted)', color='blue')
plt.title('Apple (AAPL) Stock Price Trend')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# 1. جلب بيانات تسلا والذهب معاً مع تجنب الفهارس المتعددة والمعقدة
data = yf.download(["TSLA", "GC=F"], start="2025-01-01", end="2026-06-21", multi_level_index=False)

# 2. استخراج أسعار الإغلاق وتصفية أي قيم مفقودة (عطلات نهاية الأسبوع للذهب تختلف أحياناً عن الأسهم)
prices = data['Close'].dropna()

# 3. تطبيع البيانات (تحويل بداية الاستثمار إلى أساس 100 دولار لكلا الأصلين)
normalized_prices = (prices / prices.iloc[0]) * 100

# 4. حساب العائد الإجمالي النهائي كنسبة مئوية للطباعة
final_tesla = normalized_prices['TSLA'].iloc[-1] - 100
final_gold = normalized_prices['GC=F'].iloc[-1] - 100

print(f"--- النتيجة الاستثمارية النهائية المباشرة ---")
print(f"العائد الإجمالي لـ سهم تسلا (TSLA): {final_tesla:.2f}%")
print(f"العائد الإجمالي لـ عقود الذهب (GC=F): {final_gold:.2f}%")
print("------------------------------------------")

# 5. رسم المحاكاة والمقارنة على نفس اللوحة البيانية
plt.figure(figsize=(12, 6))
plt.plot(normalized_prices['TSLA'], label='Tesla (TSLA) Growth', color='red', lw=1.5)
plt.plot(normalized_prices['GC=F'], label='Gold (GC=F) Growth', color='gold', lw=2)

# إضافة خط أفقي يمثل نقطة البداية (100 دولار) للمقارنة البصرية
plt.axhline(100, color='gray', linestyle='--', alpha=0.7, label='Initial $100 Investment')

plt.title('Investment Growth Comparison: Tesla vs Gold (Base $100)', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Portfolio Value ($)', fontsize=12)
plt.legend(loc='upper left')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np

# 1. جلب بيانات سهم تسلا وتجهيزها
ticker = "TSLA"
data = yf.download(ticker, start="2025-01-01", end="2026-06-21", multi_level_index=False)

# 2. حساب المتوسطات المتحركة (Moving Averages)
data['Fast_MA'] = data['Close'].rolling(window=9).mean()   # المتوسط السريع
data['Slow_MA'] = data['Close'].rolling(window=21).mean()  # المتوسط البطيء

# 3. خوارزمية إشارات التداول (Logic)
# نضع 1 إذا كان السريع أعلى من البطيء (شراء) و 0 إذا كان العكس (بيع)
data['Signal'] = np.where(data['Fast_MA'] > data['Slow_MA'], 1, 0)
# حساب التغير في الإشارة لمعرفة نقطة التحول بالضبط (1: شراء، -1: بيع)
data['Position'] = data['Signal'].diff()

# 4. رسم السعر مع المتوسطات وإشارات البيع والشراء
plt.figure(figsize=(14, 7))
plt.plot(data['Close'], label='Tesla Price', color='black', alpha=0.5)
plt.plot(data['Fast_MA'], label='Fast MA (9 Days)', color='blue', linestyle='--')
plt.plot(data['Slow_MA'], label='Slow MA (21 Days)', color='orange', linestyle='--')

# تحديد نقاط الشراء باللون الأخضر (مثلث للأعلى)
plt.scatter(data[data['Position'] == 1].index,
            data['Close'][data['Position'] == 1],
            label='BUY Signal 🟢', marker='^', color='green', s=150, zorder=5)

# تحديد نقاط البيع باللون الأحمر (مثلث للأسفل)
plt.scatter(data[data['Position'] == -1].index,
            data['Close'][data['Position'] == -1],
            label='SELL Signal 🔴', marker='v', color='red', s=150, zorder=5)

# تحسين مظهر الرسم
plt.title('Tesla (TSLA) Trading Strategy - Moving Average Crossover', fontsize=14, fontweight='bold')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.legend(fontsize=11, loc='upper left')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
